## Случай 1: phi = π/4

#### **1. Инициализация параметров**

```
scilab

s = 6.5;      // начальное расстояние между катером и лодкой (км)
n = 2.6;      // отношение скорости катера к скорости лодки (vk/vl = 2.6)
phi = %pi/4;  // угол направления движения лодки (45 градусов)
```

**Исходные данные задачи:**


1.   В начальный момент лодка находится на расстоянии 6.5 км от катера
2.   Катер в 2.6 раза быстрее лодки
3.   Лодка движется под углом 45° к линии первоначального визирования


#### 2. Определение дифференциального уравнения


```
scilab

function dr = f(theta, r)
    dr = -n * r * (cos(theta - phi) - 1/n) / sin(theta - phi);
endfunction

```

**Создается функция, описывающая изменение расстояния r между катером и лодкой в зависимости от угла θ:**

1.   r - текущее расстояние между катером и лодкой
2.   θ - угол визирования лодки относительно начального направления
3.   Уравнение выведено из геометрии погони и кинематических соотношений


#### **3. Начальные условия**



```
scilab

theta0 = 0;   // начальный угол визирования
r0 = s;       // начальное расстояние
eps = 0.001;  // малый параметр для обхода сингулярности


```
**Задаются начальные условия и параметр для численного интегрирования:**


1.   В начальный момент угол визирования равен 0
2.   Начальное расстояние равно s = 6.5 км
3.   eps нужен для обхода точки, где знаменатель обращается в 0


#### 4. Интегрирование до сингулярности

```
scilab
theta1 = theta0:0.001:phi-eps;  // интервал углов от 0 до φ-ε
r1 = ode(r0, theta0, theta1, f); // решение ОДУ на первом интервале
```
**Решается дифференциальное уравнение на интервале от 0 до φ-ε:**

1. ode - встроенный решатель ОДУ в Scilab
2. Интегрирование идет с шагом 0.001
3. Останавливаемся перед точкой θ = φ, где знаменатель становится равным 0

#### 5. Интегрирование после сингулярности

```
scilab
theta2 = phi+eps:0.001:2*phi;        // интервал от φ+ε до 2φ
r2 = ode(r1($), phi+eps, theta2, f); // продолжение решения
```
**Продолжается решение после точки сингулярности:**

1. r1($) - последнее значение из первого решения (начальное условие для второго интервала)
2. Интегрирование начинается сразу после особой точки
3. Продолжается до угла 2φ



#### 6. Объединение результатов

```
scilab
theta = [theta1, theta2]; // объединение интервалов углов
r = [r1, r2];            // объединение соответствующих значений расстояния
```
**Соединяются результаты двух этапов интегрирования:**

1. Получается полная траектория на всем интервале

#### 7. Построение траектории катера
```
scilab
clf;                                      // очистка графического окна
polarplot(theta, r, style=color('blue')); // траектория катера в полярных координатах
```
**Строится траектория движения катера:**

1. Полярные координаты: угол θ и радиус r
2. Синий цвет соответствует траектории катера

#### 8. Построение траектории лодки

```
scilab
t = 0:0.1:10;                   // время (в условных единицах)
x_l = s + t*cos(phi);           // координата x лодки
y_l = t*sin(phi);               // координата y лодки
plot2d(x_l, y_l, style=color('red')); // траектория лодки в декартовых координатах
```
**Строится траектория движения лодки:**

1. Лодка движется прямолинейно под углом φ
2. Координаты рассчитываются в декартовой системе
3. Красный цвет соответствует траектории лодки

#### 9. Оформление графика

```
scilab
title('Случай 1: phi = pi/4'); // заголовок
xgrid(1);                      // включение сетки
```

**Добавляются заголовок и координатная сетка.**



#### Физический смысл

**Данный код моделирует классическую задачу погони:**

1. Катер (в центре полярной системы координат) всегда направлен на лодку.
2. Лодка движется по прямой под углом 45°.
3. Траектория катера получается путем решения дифференциального уравнения.
4. Особенность при θ = φ соответствует моменту, когда катер находится точно на линии движения лодки.
5. Пересечение траекторий соответствует моменту перехвата лодки катером.

## Случай 2: φ = -π/4 (движение лодки под углом -45°)

#### 1. Задание параметров

```
scilab
phi = -%pi/4;
```

**Задается направление движения лодки: -45° (вниз вправо)**

#### 2. Определение дифференциального уравнения

```
scilab
function dr = f2(theta, r)
    dr = -n * r * (cos(theta - phi) - 1/n) / sin(theta - phi);
endfunction
```
1. Функция описывает изменение расстояния r между катером и лодкой
2. Зависит от текущего угла визирования θ

#### 3. Начальные условия

```
scilab

theta0 = 0;
r0 = s;
eps = 0.001;
theta0 = 0 - начальный угол визирования

```

1. r0 = s - начальное расстояние (6.5 км)
2. eps - параметр для обхода сингулярности

#### 4. Интегрирование до сингулярности

```
scilab

theta1 = theta0:-0.001:phi+eps;
r1 = ode(r0, theta0, theta1, f2);

```

1. Решается уравнение на интервале от 0 до φ+ε
2. Шаг -0.001 (отрицательный, т.к. φ < 0)
3. Останавливаемся перед точкой θ = φ

#### 5. Интегрирование после сингулярности
```
scilab
theta2 = phi-eps:-0.001:2*phi;
r2 = ode(r1($), phi-eps, theta2, f2);
```

1. Продолжается решение от φ-ε до 2φ
2. r1($) - последнее значение с предыдущего шага
3. Обходим особую точку

#### 6. Объединение результатов

```
scilab
theta = [theta1, theta2];
r = [r1, r2];
```
Соединяются два интервала в одну траекторию

#### 7. Построение траектории катера

```
scilab
clf;
polarplot(theta, r, style=color('green'));
```

1. Очистка графического окна
2. Построение траектории катера в полярных координатах (зеленый)

#### 8. Построение траектории лодки

```
scilab
t = 0:0.1:10;
x_l = s + t*cos(phi);
y_l = t*sin(phi);
plot2d(x_l, y_l, style=color('magenta'));
```
1. Лодка движется прямолинейно под углом -45°
2. Координаты в декартовой системе (розовый)

#### 9. Оформление графика

```
scilab
title('Случай 2: phi = -pi/4 (движение лодки под углом -45°)');
xgrid(1);
```

1. Заголовок графика с указанием случая и направления движения
2. Включение сетки

## Итог
**Моделирование задачи погони для двух случаев**

В ходе лабораторной работы была реализована математическая модель задачи погони, описывающая преследование катером береговой охраны лодки браконьеров. Модель основана на дифференциальном уравнении кривой погони в полярных координатах:

**dθ/dr=−nr((cos(θ−φ)−1/n)/(sin(θ−φ))**

где:

$r$ — расстояние между катером и лодкой

$\theta$ — угол визирования лодки

$n = 2.6$ — отношение скорости катера к скорости лодки

$\varphi$ — направление движения лодки

$s = 6.5$ км — начальное расстояние
