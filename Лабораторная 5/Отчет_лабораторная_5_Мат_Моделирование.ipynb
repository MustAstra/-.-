## Выполнение

### **Задание параметров модели**

```python
a = 0.19;
b = 0.39;
c = 0.048;
d = 0.036;
```

### **Определение правой части системы**

```python
function dx = syst(t, x)
    dx(1) = -a*x(1) + c*x(1)*x(2);
    dx(2) = b*x(2) - d*x(1)*x(2);
endfunction
```

### **Начальные условия и временной интервал**

```python
t0 = 0;
x0 = [13; 18];
t = 0:0.1:100;
y = ode(x0, t0, t, syst);
```

### **Извлечение данных**

```python
n = size(y, "c");
for i = 1:n
    хищники(i) = y(1, i);
    жертвы(i) = y(2, i);
end
```

### **Построение фазового портрета**

```python
clf();
scf(0);
plot(хищники, жертвы);
xtitle('Фазовый портрет', 'Хищники (x)', 'Жертвы (y)');
xgrid(1);
```

###  **Построение графиков численности во времени**

```python
scf(1);
subplot(2,1,1);
plot(t, хищники, 'r');
xtitle('Хищники во времени', 'Время', 'Численность');
xgrid(1);

subplot(2,1,2);
plot(t, жертвы, 'b');
xtitle('Жертвы во времени', 'Время', 'Численность');
xgrid(1);
```

### **Итог по заданию**

**Стационарное состояние (точки равновесия), не вычисленное в коде, но представляющее важный аналитический результат:**

*   $x^* = \frac{b}{d} = \frac{0.39}{0.036} \approx 10.833$
*   $y^* = \frac{a}{c} = \frac{0.19}{0.048} \approx 3.958$

Эти значения представляют собой численности хищников и жертв, при которых их популяции остаются постоянными. Вокруг этих точек происходит колебательное движение.

**Графики, которые строит программа:**

1.  **Фазовый портрет** — показывает взаимосвязь численностей хищников (x) и жертв (y) друг с другом. В данном случае мы видим замкнутую траекторию, характерную для модели Лотки-Вольтерры без внешних возмущений и саморегуляции, что указывает на периодические колебания популяций.
2.  **Графики $x(t)$ и $y(t)$** — отображают динамику каждой популяции (хищников и жертв) во времени. Они демонстрируют циклические колебания численности, где пики численности жертв предшествуют пикам численности хищников.

**Начальные условия:**

*   $x(0) = 13$ хищников
*   $y(0) = 18$ жертв

Эти начальные условия определяют конкретную траекторию в фазовом пространстве и начальные фазы колебаний популяций.

**Анализ поведения системы:**

На основе фазового портрета и временных решений можно сделать вывод, что система демонстрирует устойчивые, периодические колебания популяций хищников и жертв. Отсутствие затухания или роста колебаний указывает на консервативный характер динамики в данной идеализированной модели Лотки-Вольтерры. Популяции бесконечно следуют по замкнутым орбитам вокруг точки равновесия, их численность не стабилизируется, но и не выходит за разумные пределы, установленные начальными условиями и параметрами модели.
